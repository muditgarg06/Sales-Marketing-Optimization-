```python
import pandas as pd
data = pd.read_csv("C:\\Users\\nidhi\\Documents\\smartphones.csv")
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
df=pd.read_csv('C:\\Users\\nidhi\\Documents\\sales.csv')
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
data['Spec_Score_normalized'] = (data['Spec_Score'] - data['Spec_Score'].min()) / (data['Spec_Score'].max() - data['Spec_Score'].min())
data['Price_normalized'] = (data['Price'] - data['Price'].min()) / (data['Price'].max() - data['Price'].min())

data['ranking'] = 0.5 * data['Spec_Score_normalized'] + 0.5 * (1 - data['Price_normalized'])

data = data.sort_values(by='ranking', ascending=False)

smartphone_name = input("Enter the smartphone name: ")

# Find the rank of the specific smartphone
smartphone_rank = data[data['Smartphone'] == smartphone_name]['ranking'].values

if smartphone_rank.size > 0:
    print(f"Rank of {smartphone_name}: {smartphone_rank[0]}")
else:
    print(f"{smartphone_name} not found in the dataset.")
from sklearn.preprocessing import LabelEncoder
cat_cols = ['Date']
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])
X_train = df['Date'].values.reshape(-1, 1)
smartphone_name = input("Enter the company name: ")
y_train = df[smartphone_name].values.reshape(-1, 1)
X_test = np.array([39,40,41,42,43]).reshape(-1, 1)
#y_test = df.iloc[39:44, df.columns.get_loc(smartphone_name)].values
model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
#print(y_pred)
#rmse = np.sqrt(mean_squared_error(y_test,y_pred))

# Print the evaluation metrics
#print("Root Mean Squared Error (RMSE):", rmse)
diff = np.diff(y_pred.flatten())

# Check the trend based on the sign of the difference
if np.all(diff > 0):
    trend = 'up'
elif np.all(diff < 0):
    trend = 'down'
else:
    trend = 'mixed'

print('Trend:', trend)
trend_output = trend
average_sales=int(input("Enter the avg monthly sales: "))
scaling_factor=int(input("Enter the scaling factor: "))

quantity =int((smartphone_rank * average_sales) / scaling_factor)

if trend == 'up':
    print("no. 0f quantities: ",quantity)
elif trend == 'down'and smartphone_rank>0.7:
    print("no. 0f quantities: 10")
else:
    print("no. 0f quantities: 0" )
```
